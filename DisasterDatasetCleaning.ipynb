# Additional Analysis: Combining Disaster Dataset with Development Dataset

```python
disaster2018 = disasters2[disasters2['Year'] == 2018]
disaster2018['Country'].to_list()
```

```python
development = pd.read_csv('/Users/kaitlyncrowley/Desktop/Fall2023/DATA303/FinalProject/worlddata.csv')
development = development.rename(columns={"country": "Country"})
development = development.replace({"Country": {'Bahamas, The':'Bahamas', 'United States':'United States of America',
                                              "Korea, Dem. People's Rep.":'North Korea'}})  
development['Country'].unique()
```

```python
joinedDisasterDev = disaster2018.merge(development, how='left')
```

```python
joinedDisasterDev['Country'].unique()
```

```python
# Normalize the deaths and affected counts
joinedDisasterDev['normalizedDeaths'] = (joinedDisasterDev['Total Deaths']/joinedDisasterDev['population'])*1000000
joinedDisasterDev['normalizedAffected'] = (joinedDisasterDev['Total Affected']/joinedDisasterDev['population'])*1000000
joinedDisasterDev['normalizedInjured'] = (joinedDisasterDev['No Injured']/joinedDisasterDev['population'])*1000000
#joinedDisasterDev['date']
```
